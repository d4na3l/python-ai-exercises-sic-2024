```python
import time
import sys

# Inicializar diccionarios anidados
tabla_usuarios = [
    {
        'user_id': 29756468,
        'full_name': 'Christian Rincon',
        'password': 1234,
    },
    {
        'user_id': 12345678,
        'full_name': 'Pepito Sanchez',
        'password': 4321
    }
]

tabla_edos_de_cuenta = [
    {
        'user_id': 29756468,
        'tpo_cuenta': 'Ahorro',
        'edo_cuenta': 0.00,
    },
    {
        'user_id': 12345678,
        'tpo_cuenta': 'Corriente',
        'edo_cuenta': 9156.00,
    }
]

# Función para validar las entradas del usuario
def validate_input(prompt, tipo_dato, rango=None):
    while True:
        try:
            valor = input(prompt)

            if tipo_dato == int:  # Validamos que se pase un entero
                valor = int(valor)
            elif tipo_dato == float:  # Validamos que se pase un flotante
                valor = float(valor)

            # Validamos que el valor esté dentro del rango si es necesario
            if rango and valor not in rango:
                raise ValueError

            return valor  # Devolvemos el valor válido y salimos del bucle

        except ValueError:
            print(f"Entrada inválida. Por favor, ingrese un {tipo_dato.__name__}")

# Variable que almacena los datos del usuario para futuras referencias
usuario = {}

# Lógica de autenticación
attemps = 5

while attemps > 0:
    user_ci = validate_input('Ingrese su numero de cedula: ', int)
    user_password = validate_input('Ingrese su contraseña: ', int)

    user_found = False

    # Buscamos coincidencias de las entradas dentro del diccionario de usuarios
    for user in tabla_usuarios:
        # Comparamos los datos con los existentes dentro del diccionario
        if user_ci == user['user_id'] and user_password == user['password']:
            # Llenamos la variable usuario con los datos que hicieron match
            usuario = user
            user_found = True
            break

    if user_found:  # Si encontramos un usuario válido, salimos del bucle
        break

    attemps -= 1

    if attemps == 2:  # Penitencia de 5 segundos al equivocarse
        print(f'Error al iniciar sesión, espere 5 segundos y reintente. Intentos restantes: {attemps}')
        time.sleep(5)  # Cambiado a 5 segundos
        continue

    print(f'Error al iniciar sesión, te quedan {attemps} intentos')

# Si no hay más intentos disponibles, el programa se cerrará
if attemps == 0:
    print("Has agotado todos los intentos. El programa se cerrará.")
    sys.exit()  # Para cancelar la ejecución

# Lógica del programa
opc = 0
datos = {}

for cuenta in tabla_edos_de_cuenta:
    if cuenta['user_id'] == usuario['user_id']:
        datos = cuenta
        break

print(f'¡Bienvenido, {usuario["full_name"]}!')

while opc != 4:
    print('''Selecciona una opción:
        1. Estado de cuenta
        2. Depósito
        3. Retiro
        4. Salir
    ''')

    opc = validate_input('Ingresa opción: ', int, rango=range(1, 5))

    if opc == 1:
        print(f'Tu estado de cuenta es de: {datos["edo_cuenta"]:.2f}')

    elif opc == 2:
        print('Realizar depósito')
        deposito = validate_input('Ingrese el monto a depositar: ', float)

        confirmacion_deposito = validate_input(f'¿Seguro que quiere depositar {deposito:.2f} en su cuenta? (1=Sí, 2=No): ', int, rango=[1, 2])

        if confirmacion_deposito == 1:
            for cuenta in tabla_edos_de_cuenta:
                if datos['user_id'] == cuenta['user_id']:
                    cuenta['edo_cuenta'] += deposito
                    datos = cuenta
                    break
        else:
            print('¡Vuelva pronto!')
            opc = 0

    elif opc == 3:
        print('Realizar retiro')
        retiro = validate_input('¿De cuánto será su retiro? ', float)

        if datos['edo_cuenta'] < retiro:
            print('Saldo insuficiente, verifique su estado de cuenta')
            opc = 0
        else:
            confirmacion_retiro = validate_input(f'¿Seguro que quiere hacer un retiro de {retiro:.2f}? (1=Sí, 2=No): ', int, rango=[1, 2])
            if confirmacion_retiro == 1:
                for cuenta in tabla_edos_de_cuenta:
                    if datos['user_id'] == cuenta['user_id']:
                        cuenta['edo_cuenta'] -= retiro
                        datos = cuenta
                        break
            else:
                print('¡Vuelva pronto!')
                opc = 0

print('¡Vuelva pronto!')
```
