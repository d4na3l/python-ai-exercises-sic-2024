```python
import time
import sys

# Inicializar diccionarios anidados
tabla_usuarios = [
    {
        'user_id': 29756468,
        'full_name': 'Christian Rincon',
        'password': 1234,
    },
    {
        'user_id': 12345678,
        'full_name': 'Pepito Sanchez',
        'password': 4321
    }
]

tabla_edos_de_cuenta = [
    {
        'user_id': 29756468,
        'tpo_cuenta': 'Ahorro',
        'edo_cuenta': 0.00,
    },
    {
        'user_id': 12345678,
        'tpo_cuenta': 'Corriente',
        'edo_cuenta': 9156.00,
    }
]

# Función para validar las entradas del usuario
def validate_input(prompt, tipo_dato, rango=None):
    while True:
        try:
            valor = input(prompt)
            if tipo_dato == int:
                valor = int(valor)
            elif tipo_dato == float:
                valor = float(valor)
            if rango and valor not in rango:
                raise ValueError
            return valor
        except ValueError:
            print(f"Entrada inválida. Por favor, ingrese un {tipo_dato.__name__}")

# Función para mostrar estado de cuenta
def show_account_state(datos):
    print(f'Tu estado de cuenta es de: {datos["edo_cuenta"]:.2f}')

# Función para realizar un depósito
def make_deposit(datos):
    print('Realizar depósito')
    deposito = validate_input('Ingrese el monto a depositar: ', float)
    confirmacion_deposito = validate_input(f'¿Seguro que quiere depositar {deposito:.2f} en su cuenta? (1=Sí, 2=No): ', int, rango=[1, 2])

    if confirmacion_deposito == 1:
        datos['edo_cuenta'] += deposito
        print('Depósito realizado con éxito.')

# Función para realizar un retiro
def make_withdrawal(datos):
    print('Realizar retiro')
    retiro = validate_input('¿De cuánto será su retiro? ', float)

    if datos['edo_cuenta'] < retiro:
        print('Saldo insuficiente, verifique su estado de cuenta.')
    else:
        confirmacion_retiro = validate_input(f'¿Seguro que quiere hacer un retiro de {retiro:.2f}? (1=Sí, 2=No): ', int, rango=[1, 2])
        if confirmacion_retiro == 1:
            datos['edo_cuenta'] -= retiro
            print('Retiro realizado con éxito.')

# Función para salir del programa
def exit_program():
    print('¡Vuelva pronto!')
    sys.exit()

# Diccionario de opciones del menú principal
option_dictionary = {
    1: show_account_state,
    2: make_deposit,
    3: make_withdrawal,
    4: exit_program  # La función de salida no requiere argumento
}

# Autenticación del usuario
def authenticate_user():
    usuario = {}
    attemps = 5

    while attemps > 0:
        user_ci = validate_input('Ingrese su numero de cedula: ', int)
        user_password = validate_input('Ingrese su contraseña: ', int)

        user_found = False
        for user in tabla_usuarios:
            if user_ci == user['user_id'] and user_password == user['password']:
                usuario = user
                user_found = True
                break

        if user_found:
            return usuario
        attemps -= 1

        if attemps == 2:
            print(f'Error al iniciar sesión, espere 5 segundos y reintente. Intentos restantes: {attemps}')
            time.sleep(5)
        else:
            print(f'Error al iniciar sesión, te quedan {attemps} intentos')

    print("Has agotado todos los intentos. El programa se cerrará.")
    sys.exit()

# Programa principal
usuario = authenticate_user()
datos = next((cuenta for cuenta in tabla_edos_de_cuenta if cuenta['user_id'] == usuario['user_id']), None)
print(f'¡Bienvenido, {usuario["full_name"]}!')

opc = 0
while opc != 4:
    print('''\nSelecciona una opción:
    1. Estado de cuenta
    2. Depósito
    3. Retiro
    4. Salir
    ''')

    opc = validate_input('Ingresa opción: ', int, rango=range(1, 5))
    # Llamada a la función correspondiente en el diccionario de opciones
    if opc == 4:
        option_dictionary[opc]()  # Llama a `exit_program` sin argumentos
    else:
        option_dictionary[opc](datos)  # Pasa `datos` como argumento a las demás funciones
```
